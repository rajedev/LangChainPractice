```python
from langchain_core.messages import HumanMessage
from langchain_ollama import ChatOllama
from langchain.agents import create_agent


llm = ChatOllama(model="llama3.2:latest", temperature=0)
```

```python
agent = create_agent(model=llm, system_prompt="You are an AI assistant help us on the user questions")
response = agent.invoke({
    "messages":[
        HumanMessage(content="Tell me details on python jupiter notebook in 2 lines?")
    ]
}
)
print(response["messages"][-1])
#result = llm.invoke("Tell me details on python jupiter notebook in 2 lines?")
#result.content_blocks
```
